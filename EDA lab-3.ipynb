```python
import matplotlib.pyplot as plt
y1=[]
y2=[]
x=range(-100,100,10)
for i in x:
    y1.append(i**2)
for i in x:
    y2.append(-i**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple graph')
plt.axhline()
plt.axvline()
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print("First 5 rows of dataset:")
print(df.head())
skewness_value=df.skew()
print("Skewness value:")
print(skewness_value)
kurtosis_value=df.kurtosis()
print("Kurtosis value:")
print(kurtosis_value)
```
