```python
import numpy as np
from scipy.spatial import distance

pointA=np.array([2,4,6])
pointB=np.array([5,1,9])

#Euclidean Distance
ed=distance.euclidean(pointA,pointB)
print("Euclidean Distance: ",ed)

#Similarity (inverse of distance)
s_ed=1/(1+ed)
print("Euclidean Similarity: ",s_ed)

#Manhattan (city block) Distance
Manhattan=distance.cityblock(pointA,pointB)
print("Manhattan Distance: ",Manhattan)

#Similarity (inverse of distance)
s_man=1/(1+Manhattan)
print("Manhattan Similarity: ",s_man)

#Minckowki Distance(generalized form)
minckowki=distance.minkowski(pointA,pointB,3)
print("Minckowki Distance: (p=3) ",minckowki)

#similarity (inverse of distnace)
s_min=1/(1+minckowki)
print("Minckowki Similarity (p=3): ",s_min)

#p=2
minckowki2=distance.minkowski(pointA,pointB,2)
print("Minckowki Distance (p=2): ",minckowki2)
s_min=1/(1+minckowki2)
print("Minckowki Similarity(p=2): ",s_min)
```

```python

def hamming_distance(str1,str2):

  #Ensure strins are of equal length
  if len(str1) != len(str2):
    raise ValueError('String must be of eqaul length')

  #Count differing positions
  return sum(ch1!=ch2 for ch1,ch2 in zip(str1,str2))

#Example usage
s1='Cherihma'
s2='Cherishm'

distance=hamming_distance(s1,s2)
print("Hamming distance between",s1,"and",s2,"is",distance)
```

```python

#Token based measures(Sets/text)

def jaccard_index(str1,str2):
  set1, set2=set(str1.split()),set(str2.split())
  intersection=set1.intersection(set2)
  union=set1.union(set2)
  return len(intersection) / len(union)

s1="data science is fun"
s2='science makes data useful'

print("Jaccard index:",jaccard_index(s1,s2))
```

```python

#Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1="data science is fun"
s2="science makes data useful"

vectorizer = CountVectorizer().fit([s1,s2])
vectors=vectorizer.transform([s1,s2])

cos_sim = cosine_similarity(vectors[0],vectors[1])[0][0]
print("Cosine similarity:",cos_sim)

```

```python
#Sequence-based measures
def lcs_length(X,Y):
  m,n=len(X),len(Y)
  dp=[[0]*(n+1) for _ in range(m+1)]

  for i in range(m):
    for j in range(n):
      if X[i]==Y[j]:
        dp[i+1][j+1]=dp[i][j]+1
      else:
        dp[i+1][j+1]=max(dp[i][j+1],dp[i+1][j])
  return dp[m][n]

seq1="aabbaa"
seq2="aaaabba"

length=lcs_length(seq1,seq2)
print("Length of LCS:",length)
```
