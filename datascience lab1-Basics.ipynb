```python
#largest of three numbers
a=int(input('enter value a'))
b=int(input('enter value b'))
c=int(input('enter value c'))
if(a>b and a>c):
  print('a is the largest number')
elif(b>a and b>c):
  print('b is the largest number')
else:
  print('c is the largest number')
```
